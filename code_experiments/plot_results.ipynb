```python
from pathlib import Path
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

prjpath = Path("..")


modelsfolder = Path(prjpath, "models")
datafolder = Path(prjpath, "data")
resfolder = Path(prjpath, "results")
figfolder = Path(prjpath, "figures")
```

```python
filename = Path(resfolder, "cancer_procdata_results20240831-111624.csv")
df = pd.read_csv(filename, index_col=0)
```

```python
Path(figfolder,"results-pn.pdf")
```

```python
sns.histplot(data=df, x="pn", )
plt.xlabel('$PN(HN, D)$')
plt.xlim(df.pn.min()*0.999, 1)
plt.rcParams.update({'font.size': 12})
plt.savefig(Path(figfolder,"results-pn.pdf"), format="pdf")


# Show the plot
plt.show()
```

```python
import matplotlib.ticker as ticker

plot_ = sns.histplot(data=df, x="ps", )
plt.xlabel('$PS(HN, D)$')
plt.xlim(0,df.ps.max()*1.1)
plot_.xaxis.set_major_locator(ticker.LinearLocator(5))
plt.rcParams.update({'font.size': 12})

plt.savefig(Path(figfolder,"results-ps.pdf"), format="pdf")

# Show the plot
plt.show()
```
